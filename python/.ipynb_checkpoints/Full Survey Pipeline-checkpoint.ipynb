```python
# Load Libraries
import pandas as pd
import numpy as np
```

```python
# Import Data Files (Student Survey & School Enrollment Files)
df_studentsurvey = pd.read_csv('/Users/yuriyang/Desktop/Scholarfund/full_student_survey_data.csv')
df_schoolenrollment = pd.read_csv('/Users/yuriyang/Desktop/Scholarfund/school_enrollment_data copy.csv')
```

```python
# Clean Student Dataframe: only select the columns that are relevant
# (You should only have to run this step if you used the preset dataframe names)

df_studentsurvey_subset = df_studentsurvey[[
    'Response ID',
    'Status',
    'Language',
    'School Code',
    'coppa_age',
    'q_aspire_self',
    'q_expect_self',
    'q_expect_family',
    'q_expect_friends',
    'q_expect_staff',
    'q_grad_ready',
    'q_teachers_expect_all',
    'q_teachers_high_expectations',
    'q_teachers_expect_hardwork',
    'q_principal_future_plan',
    'q_school_college_match',
    'q_school_career_match',
    'q_learning_career',
    'q_school_achieve_career',
    'q_develop_financial',
    'q_hsbp_useful',
    'q_hsbp_platform_confidence',
    'q_hsbp_more_likely',
    'q_college_future_career',
    'q_school_dcoptions',
    'q_know_reqs_graduate',
    'q_know_reqs_apprenticeship',
    'q_know_reqs_2yr',
    'q_know_reqs_4yr',
    'q_know_reqs_posths',
    'q_staff_ontrack',
    'q_staff_support',
    'q_school_1on1supp',
    'q_staff_met_college',
    'q_advisory_plan_posths',
    'Parents / Guardians:q_most_helpful',
    'Other family members (siblings, cousins, uncle/aunt):q_most_helpful',
    'Teachers:q_most_helpful',
    'Counselors:q_most_helpful',
    'College and Career Specialist:q_most_helpful',
    'Other school staff:q_most_helpful',
    'Myself / my own research:q_most_helpful',
    'Friends / other community members:q_most_helpful',
    'q_satisfied_cc',
    'q_school_parent_welcome',
    'q_school_adult_knows_well',
    'q_school_adult_problem',
    'q_finaid_complete',
    'q_finaid_purpose',
    'q_school_talktofam',
    'q_afford_college',
    'q_cost_4yr',
    'q_cost_2yr',
    'q_school_talk_scholarships',
    'q_parents_edlevel'
]]
```

```python
# Clean School dataframe: only select the School Code & School Name Columns
df_schoolenrollment_subset = df_schoolenrollment[['SchoolCode', 'SchoolName']]
```

```python
# Clean Student dataframe: only pull data for responses where Status is 'Complete' and coppa_age is equal to 2
df_studentsurvey_filtered = df_studentsurvey_subset[
    (df_studentsurvey_subset['Status'] == 'Complete') &
    (df_studentsurvey_subset['coppa_age'] == 2)
]

#df_studentsurvey_filtered = df_studentsurvey_subset
```

```python
# Clean Student & School dataframes: Rename columns
# Please use these naming conventions for the column names
  # Student Data Frame
    #'Response ID':'response_id'
    #'Status':'status'
    #'Language':'language'
    #'School Code':'school_code'
    #'Parents / Guardians:q_most_helpful':'q_most_helpful_parents'
    #'Other family members (siblings, cousins, uncle/aunt):q_most_helpful':'q_most_helpful_otherfam'
    #'Teachers:q_most_helpful':'q_most_helpful_teachers'
    #'Counselors:q_most_helpful':'q_most_helpful_counselors'
    #'College and Career Specialist:q_most_helpful':'q_most_helpful_ccspecialist'
    #'Other school staff:q_most_helpful':'q_most_helpful_otherschoolstaff'
    #'Myself / my own research:q_most_helpful':'q_most_helpful_myself'
    #'Friends / other community members:q_most_helpful':'q_most_helpful_friends'

df_studentsurvey_filtered = df_studentsurvey_filtered.rename(columns={
    'Response ID': 'response_id',
    'Status': 'status',
    'Language': 'language',
    'School Code': 'school_code',
    'Parents / Guardians:q_most_helpful': 'q_most_helpful_parents',
    'Other family members (siblings, cousins, uncle/aunt):q_most_helpful': 'q_most_helpful_otherfam',
    'Teachers:q_most_helpful': 'q_most_helpful_teachers',
    'Counselors:q_most_helpful': 'q_most_helpful_counselors',
    'College and Career Specialist:q_most_helpful': 'q_most_helpful_ccspecialist',
    'Other school staff:q_most_helpful': 'q_most_helpful_otherschoolstaff',
    'Myself / my own research:q_most_helpful': 'q_most_helpful_myself',
    'Friends / other community members:q_most_helpful': 'q_most_helpful_friends'
})

 # School Data Frame
    #'SchoolCode':'school_code',
    #'SchoolName':'school_name',

df_schoolenrollment_subset = df_schoolenrollment_subset.rename(columns={
    'SchoolCode': 'school_code',
    'SchoolName': 'school_name'
})
```

```python
# Create Most Helpful Measure (q_most_helpful) in Student dataframe 
# Logic: 
    # You should use the case_when function
    # You will need to use the 8 'q_most_helpful_*' columns
        # q_most_helpful_parents
        # q_most_helpful_otherfam
        # q_most_helpful_teachers
        # q_most_helpful_counselors
        # q_most_helpful_ccspecialist
        # q_most_helpful_otherschoolstaff
        # q_most_helpful_myself
        # q_most_helpful_friends
    # The numeric value shows what rank students gave this particular group with 1 being the most helpful and 8 being the least helpful
    # q_most_helpful should only be set with a single text/string value that corresponds with the respective columns
        # Parents / Guardians
        # Other family members (siblings, cousins, uncle/aunt)
        # Teachers
        # Counselors
        # College and Career Specialist
        # Other school staff
        # Myself / my own research
        # Friends / other community members

# First step is to create the column and set it initially with 'Unknown' for all values
df_studentsurvey_filtered['q_most_helpful'] = 'Unknown'

# Enter the code for the case_when function below
conditions = [
    df_studentsurvey_filtered['q_most_helpful_parents'] == 1,
    df_studentsurvey_filtered['q_most_helpful_otherfam'] == 1,
    df_studentsurvey_filtered['q_most_helpful_teachers'] == 1,
    df_studentsurvey_filtered['q_most_helpful_counselors'] == 1,
    df_studentsurvey_filtered['q_most_helpful_ccspecialist'] == 1,
    df_studentsurvey_filtered['q_most_helpful_otherschoolstaff'] == 1,
    df_studentsurvey_filtered['q_most_helpful_myself'] == 1,
    df_studentsurvey_filtered['q_most_helpful_friends'] == 1
]

choices = [
    'Parents / Guardians',
    'Other family members (siblings, cousins, uncle/aunt)',
    'Teachers',
    'Counselors',
    'College and Career Specialist',
    'Other school staff',
    'Myself / my own research',
    'Friends / other community members'
]

df_studentsurvey_filtered['q_most_helpful'] = np.select(conditions, choices, default='Unknown')
```

```python
df_studentsurvey_filtered[['q_most_helpful','q_most_helpful_parents', 'q_most_helpful_otherfam', 'q_most_helpful_teachers', 'q_most_helpful_counselors', 'q_most_helpful_ccspecialist', 'q_most_helpful_otherschoolstaff', 'q_most_helpful_myself', 'q_most_helpful_friends']].head(10)
```

```python
# Create First Gen Dimension in student Data Frame
#df_studentsurvey_filtered['first_gen'] = 0
#df_studentsurvey_filtered.loc[df_studentsurvey_filtered['q_parents_edlevel'] == 1, 'first_gen'] = 1

df_studentsurvey_filtered['first_gen'] = 2 - df_studentsurvey_filtered['q_parents_edlevel']
```

```python
# link Student Survey dataframe to School Name data frame using a left join
df_linked = df_studentsurvey_filtered.merge(
    df_schoolenrollment_subset,
    how='left',
    on='school_code')
```

```python
# Drop the 8 Columns that were used for creating the Most Helpful measure/column and the coppa_age & Status columns
columns_to_drop = [
        # q_most_helpful_parents
        # q_most_helpful_otherfam
        # q_most_helpful_teachers
        # q_most_helpful_counselors
        # q_most_helpful_ccspecialist
        # q_most_helpful_otherschoolstaff
        # q_most_helpful_myself
        # q_most_helpful_friends
    'q_most_helpful_parents',
    'q_most_helpful_otherfam',
    'q_most_helpful_teachers',
    'q_most_helpful_counselors',
    'q_most_helpful_ccspecialist',
    'q_most_helpful_otherschoolstaff',
    'q_most_helpful_myself',
    'q_most_helpful_friends',
    'coppa_age',
    'status'
]

df_linked = df_linked.drop(columns=columns_to_drop, axis = 1)
```

```python
df_linked.head(50)
```

```python
# Unpivot Data using the melt function
    # Response ID
    # Language
    # School Code
    # School Name
    # First Generation Status

df_data_long = df_linked.melt(id_vars = ['response_id','language', 'school_code', 'school_name', 'first_gen'])
```

```python
df_data_long.head(10)
```

```python
# Create Functions that will be used to Code Text Values for each type of Survey Question

# Example Code for Yes/No Questions
def yes_no_coding(series):
    if series == 1:
        return 'No'
    elif series == 2:
        return 'Yes'
    else:
        return 'REVIEW'
    
# Placeholder Code for Agree/Disagree Questions
def agree_coding(series): 
    if series == 1:
        return 'Strongly Disagree'
    elif series == 2:
        return 'Disagree'
    elif series == 3:
        return 'Agree'
    elif series == 4:
        return 'Strongly Agree'
    else:
        return 'REVIEW'
        
# Enter Code for Education Level Questions below
def edu_level(series):
    if series == 1:
        return 'High School/Straight to Career or Military'
    elif series == 2:
        return 'Apprenticeship'
    elif series == 3:
        return 'Some College'
    elif series == 4:
        return '2-Year College'
    elif series == 5:
        return '4-Year College'
    elif series == 6:
        return 'Graduate School'
    else:
        return 'REVIEW'

# Enter Code for Satisfied Questions below
def satisfied(series): 
    if series == 1:
        return 'Very Unsatisfied'
    elif series == 2:
        return 'Unsatisfied'
    elif series == 3:
        return 'Satisfied'
    elif series == 4:
        return 'Very Satisfied'
    else:
        return 'REVIEW'

# Enter Code for Cost Questions below    
def cost(series):
    if series == 1:
        return '$0 - $5,000'
    elif series == 2:
        return '$5,001 - $10,000'
    elif series == 3:
        return '$10,001 - $15,000'
    elif series == 4:
        return '$15,001 - $20,000'
    elif series == 5:
        return '$20,001 - $25,000'
    elif series == 6:
        return '$25,001 - $30,000'
    elif series == 7:
        return '$30,001 - $35,000'
    elif series == 8:
        return '$35,001 - $40,000'
    elif series == 9:
        return '$40,001 - $45,000'
    elif series == 10:
        return '$45,001 - $50,000'
    elif series == 11:
        return 'Over $50,000'
    else:
        return 'REVIEW'
```

```python
# Initially set the text_value column to equal the value column for all rows
df_data_long['text_value'] = df_data_long['value'] 
```

```python
df_data_long.head(10)
```

```python
# Create a new data frame that only includes survey questions that were Yes/No Questions
# Then apply the appropriate function to Code Text Values to that subset of questions

df_data_long_yesno = df_data_long.loc[df_data_long['variable'].isin(['q_expect_family','q_expect_friends','q_expect_staff',
                                                                     'q_grad_ready','q_teachers_expect_all','q_teachers_high_expectations',
                                                                     'q_teachers_expect_hardwork','q_principal_future_plan',
                                                                     'q_school_dcoptions', 'q_staff_met_college','q_finaid_complete',
                                                                     'q_finaid_purpose','q_school_talktofam','q_afford_college',
                                                                     'q_school_talk_scholarships', 'q_parents_edlevel'
                                                                    ])].reset_index()

df_data_long_yesno['text_value'] = df_data_long_yesno['text_value'].apply(yes_no_coding)
```

```python
df_data_long_yesno.head(10)
```

```python
# Create a new data frame that only includes survey questions that were Agree/Disagree Questions
# Then apply the appropriate function to Code Text Values to that subset of questions
df_data_long_agree = df_data_long.loc[df_data_long['variable'].isin([
    'q_school_college_match', 'q_school_career_match', 'q_learning_career','q_school_achieve_career',
    'q_develop_financial','q_hsbp_useful','q_hsbp_platform_confidence','q_hsbp_more_likely',
    'q_college_future_career','q_know_reqs_graduate','q_know_reqs_apprenticeship','q_know_reqs_2yr',
    'q_know_reqs_4yr','q_know_reqs_posths','q_staff_ontrack','q_staff_support','q_school_1on1supp',
    'q_advisory_plan_posths','q_school_parent_welcome','q_school_adult_knows_well','q_school_adult_problem'
])].reset_index()

df_data_long_agree['text_value'] = df_data_long_agree['text_value'].apply(agree_coding)
```

```python
df_data_long_agree.head(10)
```

```python
# Create a new data frame that only includes survey questions that were Education Level Questions
# Then apply the appropriate function to Code Text Values to that subset of questions
df_data_long_edu = df_data_long.loc[df_data_long['variable'].isin(['q_aspire_self','q_expect_self'])].reset_index()

df_data_long_edu['text_value'] = df_data_long_edu['text_value'].apply(edu_level)
```

```python
df_data_long_edu.head(10)
```

```python
# Create a new data frame that only includes survey questions that were Satisfied Questions
# Then apply the appropriate function to Code Text Values to that subset of questions
df_data_long_satisfited = df_data_long.loc[df_data_long['variable'].isin(['q_satisfied_cc'])].reset_index()

df_data_long_satisfited['text_value'] = df_data_long_satisfited['text_value'].apply(satisfied)
```

```python
df_data_long_satisfited.head(10)
```

```python
# Create a new data frame that only includes survey questions that were Cost Questions
# Then apply the appropriate function to Code Text Values to that subset of questions
df_data_long_cost = df_data_long.loc[df_data_long['variable'].isin(['q_cost_4yr','q_cost_2yr'])].reset_index()

df_data_long_cost['text_value'] = df_data_long_cost['text_value'].apply(cost)
```

```python
df_data_long_cost.head(10)
```

```python
# Create a new data frame for the Most Helpful question
# Code the value field to Not a Number (np.nan) for all rows
df_data_long_most_helpful = df_data_long.loc[df_data_long['variable'] == 'q_most_helpful'].reset_index()
df_data_long_most_helpful['value'] = np.nan
```

```python
df_data_long_most_helpful.head(10)
```

```python
# Union the 6 data frames above back together into a single field
df_union = pd.concat([df_data_long_yesno
                      ,df_data_long_agree
                      ,df_data_long_edu
                      ,df_data_long_satisfited
                      ,df_data_long_cost
                      ,df_data_long_most_helpful])
```

```python
# Check that the row count of the original dataframe matches the row count for the new dataframe
len(df_union) == len(df_data_long)
```
